```python
from pydantic import BaseModel
from typing import List, Optional

class Ingredient(BaseModel):
    name: str
    quantity: str
    substitute: Optional[str]

class Step(BaseModel):
    step_number: int
    instruction: str

class Recipe(BaseModel):
    dish_name: str
    ingredients: List[Ingredient]
    steps: List[Step]
    cooking_time: str
    difficulty: str
```
